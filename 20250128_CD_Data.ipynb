```python

from google.colab import files
import numpy as np
import pandas as pd
from tkinter import Tk     # from tkinter import Tk for Python 3.x
from tkinter.filedialog import askopenfilename
import matplotlib.pyplot as plt
import os
import time
from google.colab import files
import io
# Upload the CSV file
uploaded = files.upload()
filename = next(iter(uploaded))

# Read the CSV file into a pandas DataFrame
df = pd.read_excel(io.BytesIO(uploaded[filename]))

# Now you can work with the DataFrame 'df'
print(df)

rows1 = len(df.index)
```

```python
#Averaging the Data

columns_to_average = ['7_96_1', '7_96_2', '7_96_3']
# Calculate the average and add it as a new column
df['7_96_avg'] = df[columns_to_average].mean(axis=1)
# repeat for all columns
columns_to_average = ['7_51_1', '7_51_2', '7_51_3']
df['7_51_avg'] = df[columns_to_average].mean(axis=1)
columns_to_average = ['7_03_1', '7_03_2', '7_03_3']
df['7_03_avg'] = df[columns_to_average].mean(axis=1)
columns_to_average = ['6_55_1', '6_55_2', '6_55_3']
df['6_55_avg'] = df[columns_to_average].mean(axis=1)
columns_to_average = ['6_03_1', '6_03_2', '6_03_3']
df['6_03_avg'] = df[columns_to_average].mean(axis=1)
columns_to_average = ['5_54_1', '5_54_2', '5_54_3']
df['5_54_avg'] = df[columns_to_average].mean(axis=1)
columns_to_average = ['5_14_1', '5_14_2', '5_14_3']
df['5_14_avg'] = df[columns_to_average].mean(axis=1)


# Print the updated DataFrame
print(df)
```

```python
# prompt: smooth the data provided using savitsky-golay filter

from scipy.signal import savgol_filter

# Assuming your data is in a column named 'column_name' in the DataFrame 'df'
# Replace 'column_name' with the actual column name

# Example: Assuming the data is in the first column
column_name = df.columns[0]
smoothed_7_96 = savgol_filter(df["7_96_avg"], window_length=5, polyorder=2)

# Add the smoothed data as a new column to the DataFrame
df['7_96_smooth'] = smoothed_7_96

# Print the updated DataFrame
print(df)


# Plot the original and smoothed data
plt.figure(figsize=(10, 6))
plt.plot(df['wavelength'], df["7_96_avg"], label='Original Data')
plt.plot(df['wavelength'], df['7_96_smooth'], label='Smoothed Data')
plt.xlabel('wavelength')
plt.ylabel('Values')
plt.title('Original vs. Smoothed Data')
plt.legend()
plt.show()
```

```python
# Calculate residuals
residuals = df['7_96_avg'] - df['7_96_smooth']

# Create the residual plot
plt.figure(figsize=(10, 6))
plt.plot(df['wavelength'], residuals, label='Residuals')
plt.axhline(y=0, color='r', linestyle='--', label='Zero Line')  # Add a zero line for reference
plt.xlabel('Wavelength')
plt.ylabel('Residuals')
plt.title('Residual Plot for Savitzky-Golay Smoothing')
plt.legend()
plt.show()
```

```python
# Smoothing all averaged data points

df['7_51_smooth'] = savgol_filter(df["7_51_avg"], window_length=5, polyorder=2)
df['7_03_smooth'] = savgol_filter(df["7_03_avg"], window_length=5, polyorder=2)
df['6_55_smooth'] = savgol_filter(df["6_55_avg"], window_length=5, polyorder=2)
df['6_03_smooth'] = savgol_filter(df["6_03_avg"], window_length=5, polyorder=2)
df['5_54_smooth'] = savgol_filter(df["5_54_avg"], window_length=5, polyorder=2)
df['5_14_smooth'] = savgol_filter(df["5_14_avg"], window_length=5, polyorder=2)
print(df)

```
