```python
# prompt: a bar chart

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["WT", "4.3", "4.13", "4.17", "4.21", "4.22", "4.23", "4.24"])
y = np.array([0.0075, 0.0051, 0.0054, 0.0039, 0.0045, 0.015, 0.0045, 0.0051])
errors = np.array([0.0008, 0.0015, 0.0006, 0.0006, 0.0010, 0.0089, 0.0003, 0.0005])

bars = plt.bar(x,y, yerr=errors, error_kw={'capsize': 8, 'elinewidth': 1, 'ecolor': 'gray'}, color = '#21918c', edgecolor = 'black')

plt.title("Activity of some round 4 mutants")
plt.xlabel("Variant")
plt.ylabel("Mean turnover (per hour)")

bars[5].set_color('#5ec962')
bars[5].set_edgecolor('black')
bars[0].set_color('#3b528b')
bars[0].set_edgecolor('black')

plt.show()
```

```python
import matplotlib.cm as cm

# Choose a colormap (e.g., 'viridis')
cmap = cm.get_cmap('viridis', len(x))

# Create the bar chart
bars = plt.bar(x, y, yerr=errors)

# Set colors for each bar
for i, bar in enumerate(bars):
    bar.set_color(cmap(i))


plt.show()
```

```python
plt.figure(figsize=(5, 3), dpi=240)

bars = plt.bar(x,y, yerr=errors, error_kw={'capsize': 2, 'elinewidth': 1, 'ecolor': 'black'}, edgecolor = 'black')

plt.title("Activities of Selected DE Round 4 Mutants")
plt.xlabel("Variant")
plt.ylabel("Mean turnover (per hour)")

# Set colors for each bar
for i, bar in enumerate(bars):
    bar.set_color(cmap(i))

bars[0].set_edgecolor('black')
bars[1].set_edgecolor('black')
bars[2].set_edgecolor('black')
bars[3].set_edgecolor('black')
bars[4].set_edgecolor('black')
bars[5].set_edgecolor('black')
bars[6].set_edgecolor('black')
bars[7].set_edgecolor('black')

plt.show()
```
