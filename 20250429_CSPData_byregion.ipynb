```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming you have already defined regions, pH_vals, and df_csp

# Get the data for each region from df_csp
region_data = []
for region in regions:
    csp_values = df_csp.loc[df_csp['Region'] == region].drop(columns=['Region']).values.flatten().tolist()
    region_data.append(csp_values)

# Set the width of each bar
bar_width = 0.8 / len(regions)  # Adjust as needed for spacing

# Create the bar chart
fig, ax = plt.subplots(figsize=(10, 6))  # Adjust figure size if needed

# Create bars for each region
for i, region in enumerate(regions):
    x_positions = [x + i * bar_width for x in range(len(pH_vals))]
    stdev_values = df_stdev.loc[df_stdev['Region'] == region].drop(columns=['Region']).values.flatten().tolist()
    ax.bar(x_positions, region_data[i], width=bar_width, label=region, yerr=stdev_values, capsize=2)

# Customize the plot
ax.set_xlabel('pH')
ax.set_ylabel('CSP (ppm)')
ax.set_title('CSP of Regions vs. pH')
ax.set_xticks([r + bar_width * (len(regions) / 2 - 0.5) for r in range(len(pH_vals))])
ax.set_xticklabels(pH_vals)
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Place legend outside the plot
plt.tight_layout()  # Adjust layout to prevent labels from overlapping

# Display the plot
plt.show()
```
